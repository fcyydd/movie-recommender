```python
import pandas as pd
```

```python
data = pd.read_csv("Dataset/imdb(MOVIES)_processed.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
#Delete spaces from columns
columns = ['Director', 'Actor1', 'Actor2', 'Actor3', 'Actor4']
dct = {
    'Director': [],
    'Actor1': [],
    'Actor2': [],
    'Actor3': [],
    'Actor4': []
}

for index, row in data.iterrows():
    director = data['Director']
    for col in columns:
        word = ''
        word += ''.join(row[col]).replace(' ', '')
        dct[col].append(word)

for col in columns:
    data[col] = dct[col]

data.head()
```

```python
#Create a column of word representation by combining column attributes
import re

data['Bag of Words'] = ''
columns = ['Genre', 'Director', 'Actor1', 'Actor2', 'Actor3', 'Actor4', 'Keywords']
bag_of_words = []
for index, row in data.iterrows():
    words = ''
    for col in columns:
        words += ''.join(row[col]).lower() + ' '
        words = re.sub(r'[\[\'\]\,\.]', '' , words)
    bag_of_words.append(words)
    
data['Bag of Words'] = bag_of_words
data.head()
```

```python
#Vector representation for 'Bag of Words' and the similarity matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
count_matrix = count.fit_transform(data['Bag of Words'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
# Create a Series for 'Movie Name'
indices = pd.Series(data['Movie Name'])
indices[:5]
```

```python
# Create recommender model
def recommend(title, cosine_sim=cosine_sim):
    recommend_movies = []
    idx = indices[indices == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top_10 = list(score_series.iloc[1:11].index)

    for i in top_10:
        recommend_movies.append(list(data['Movie Name'])[i])

    return recommend_movies
```

```python
recommend('The Matrix')
```

```python
recommend('It Happened One Night')
```
